```python
# Problem 1: Smart Parking Lot Management System

def parking_system(capacity, logs):
    parked = 0
    peak = 0

    for log in logs:
        if log == "IN":
            parked += 1
        elif log == "OUT" and parked > 0:
            parked -= 1
        peak = max(peak, parked)

    status = "Full" if parked >= capacity else "Available"

    print("Currently Parked Vehicles:", parked)
    print("Peak Usage:", peak)
    print("Parking Status:", status)

parking_system(50, ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"])
```

```python
# Problem 2: Online Food Delivery Time Estimator
def delivery_time(distance, traffic, weather):
    eta = distance * 5  # base time: 5 minutes per km

    if traffic == "High":
        eta += 15
    elif traffic == "Medium":
        eta += 10

    if weather == "Rainy":
        eta += 10
    elif weather == "Stormy":
        eta += 20

    print("Estimated Delivery Time:", eta, "minutes")

delivery_time(8, "High", "Rainy")
```

```python
# Problem 3: Movie Theatre Seat Occupancy Analyzer
def seat_occupancy(total_seats, booked_seats):
    booked = len(booked_seats)
    occupancy = (booked / total_seats) * 100

    if occupancy == 100:
        status = "Housefull"
    elif occupancy >= 75:
        status = "Almost Full"
    else:
        status = "Seats Available"

    print("Occupancy:", f"{occupancy:.0f}%")
    print("Show Status:", status)

seat_occupancy(200, [1]*150)
```

```python
# Problem 4: Cloud Server Load Classification System
def server_load(cpu_readings):
    avg_load = sum(cpu_readings) / len(cpu_readings)

    if avg_load < 50:
        status = "Normal"
    elif avg_load <= 80:
        status = "Warning"
    else:
        status = "Critical"

    print("Average CPU Load:", f"{avg_load:.0f}%")
    print("Server Status:", status)

server_load([45, 60, 70, 85, 90])
```

```python
# Problem 5: Smart Classroom Resource Usage Monitor
def resource_monitor(usage):
    overused = [resource for resource, hours in usage.items() if hours > 8]

    print("Overused Resources:", ", ".join(overused) if overused else "None")
    print("Energy Alert:", "Yes" if overused else "No")

resource_monitor({
    "Projector": 6,
    "AC": 9,
    "Lights": 4
})
```

```python
# Problem 6: Online Event Registration Capacity Controller
def event_registration(capacity, registrations):
    confirmed = min(capacity, registrations)
    waitlist = max(0, registrations - capacity)
    status = "Closed" if registrations >= capacity else "Open"

    print("Confirmed Registrations:", confirmed)
    print("Waitlisted Users:", waitlist)
    print("Registration Status:", status)

event_registration(100, 105)
```
